The Matrix Algebra

```python
import numpy as np

ab = np.array([2,4,5])
bc = np.array([3,4,6])

print(np.dot(ab, bc))

a = np.array([[1,2,3],[4,5,6]])
b = np.array([[1,2],[3,4],[5,6]])

adotb = np.dot(a,b)

print(adotb)
```

Used in:

similarity

attention scores

cosine similarity

```python
x = np.array([[1,2],
              [3,4]])

y = np.array([[5,6],
              [7,8]])

print(np.matmul(x,y))
print(x@y)
```

** *Shape Rule: *(m × n) @ (n × p) = (m × p)**

Otherwise → ❌ Error

```python
# TRANSPOSE - Important

mat = np.array([[1,2,3],[4,5,6]])
transpose = mat.T
print(transpose)
```

Used in:

gradients

```python
# Identity and Inverse Matrix:

Identity = np.eye(4)

print(Identity)

demo_mat = np.array([[-18,-2,-53],[4,5,-6],[70,-18,10]])

Inverse = np.linalg.inv(demo_mat) # it only exists when determinant of this matrix is NOT equals to ZERO

print(Inverse)
```

```python
# NORMS - L2 Norms (most used)

v = np.array([5,6])

norm = np.linalg.norm(v)

print(norm)
```
