Pandas = “Excel for Python”
It helps you work with tables (rows × columns), called DataFrames.

Two core objects:

Series → Single column data

DataFrame → Multiple column table

```python
# From Dictonary

import pandas as pd

data = {
    "Name":["Krish","Ram","Annu"],
    "Marks":[44,55,98]
}

df_basic = pd.DataFrame(data)

print(df_basic)
```

```python
# From lists of list

dataa = pd.DataFrame([[0,1,3,4],[4,4,4,5]], columns=["pre","main","interview","health"])

print(dataa)
```
