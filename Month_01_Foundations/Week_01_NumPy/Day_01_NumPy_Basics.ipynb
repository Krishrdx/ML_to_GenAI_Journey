```python
arr1 = np.array([[1,2,5,7,8],
                 [3,6,9,12,15]])

print("the shape of arr1", arr1.shape)
print("the dimension of arr1", arr1.ndim)
print("the size of arr1", arr1.size)
print("the data type of arr1", arr1.dtype)
```

```python
zeros = np.zeros((5,4))
ones = np.ones((4,5))
identity = np.eye(3)

print("zeros:", zeros)
print("ones:", ones)
print("identity:", identity)
```

- NUPMY better than Python Lists because
1. they are faster
2. they are fixed not flexible
3. They are vectorized

- The fixed dtype are impt for GPUs because
1. predictable Memory alignment and they dont vary with time that provides stability to the sys.
2. the computing will get faster.
3. Low Memory allocation will happen
4. Better Parllelization : All cores execute the same instruction simultaneously.

- If the reshaped dimensions are not matched it must throw an ValueError.
